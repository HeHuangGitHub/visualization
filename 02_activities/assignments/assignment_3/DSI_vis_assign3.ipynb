```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the data of Toronto Island Ferry Ticket Counts
# https://open.toronto.ca/dataset/toronto-island-ferry-ticket-counts/
# This dataset provides ferry ticket sales and ticket redemptions.
ferry = pd.read_csv('Toronto_Island_Ferry_Ticket_Counts.csv')

# The dataframe is too big. I only keep the top 1,000 column.
ferry = ferry.head(1000)

# Convert the Timestamp column into datetime format
ferry['Timestamp'] = pd.to_datetime(ferry['Timestamp'])

# Show the first few rows to inspect the data
print(ferry.head())
```

```python
# Make a trend plot to show the changes of Redemption and Sales counts of Toronto Island Ferry Ticket over time.
plt.figure(figsize=(10, 6))
plt.plot(ferry['Timestamp'], ferry['Redemption Count'], label='Redemption Count', marker='o')
plt.plot(ferry['Timestamp'], ferry['Sales Count'], label='Sales Count', marker='s')
plt.title('Trend of Redemption Count and Sales Count of Toronto Island Ferry Ticket Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Counts')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
