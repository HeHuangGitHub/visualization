# Making a basic figure with matplotlib

```python
# load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.barh(x,y)
```

```python
fig, ax = plt.subplots(figure = (5,3))
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x, y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time lets make this a reallly long title', fontdict=font1, loc='left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

Modifiying our data points

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.scatter(x, y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,y,
        marker = '*',
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x,y,marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '000000')

ax.grid(axis = 'y',
        color = 'blue',
        linewidth = 2,
        linestyle = '-.')
```

### Python Graph Gallery

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
sns.set_theme(style="darkgrid")
sns.kdeplot(df['sepal_width'], fill=True, color="r")
sns.kdeplot(df['sepal_length'], fill=True, color="b")
plt.show()
```

Aesthetic: Yes, this visualization is very concise and clear. <br>
Substantive: Yes, this visualization presents the distribution of sepal width and sepal length of the iris. <br>
Perceptual: Yes, this visualization shows normal distribution of sepal width and sepal length of the iris with different centres and spread. <br>
